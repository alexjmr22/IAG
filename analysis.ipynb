## Visual Inspection
Displaying the latest samples from each model.

```python
def show_samples(model_name):
    sample_dir = 'samples'
    # Get all matching files
    files = sorted([f for f in os.listdir(sample_dir) if f.startswith(model_name)])
    if not files:
        print(f"No samples found for {model_name}")
        return
    
    latest = files[-1]
    print(f"Showing {latest}")
    img = Image.open(os.path.join(sample_dir, latest))
    plt.figure(figsize=(12, 12))
    plt.imshow(img)
    plt.axis('off')
    plt.show()

show_samples('gan')
```
